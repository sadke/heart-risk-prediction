```python
diag_cols = [diag for diag in heart_data.columns if "DIAG" in diag]
diag_cols.remove("DIAGNOSISCODESETNAME")
diag_cols.remove("OTHERDIAGNOSES")
print(diag_cols)
```

```python
all_codes = set()

for col in diag_cols:
    all_codes |= set(heart_data[col].dropna().unique().tolist())
print(all_codes)
```

```python
len(all_codes)
```

```python
type(list(all_codes)[0])
```

```python
from collections import defaultdict, Counter

counter = Counter()
unique_dict = defaultdict(set)

for code in list(all_codes):
    counter[code] += 1
    unique_dict[code.lower()].add(code)

print(len(unique_dict)) 
```
